# **MATMUL**

```python
mat_a = np.array([[1,2],[3,4]])
mat_b = np.array([[5,6],[7,8]])
mat_a_det = np.linalg.det(mat_a)
mat_b_det = np.linalg.det(mat_b)
```

```python
mat_a_det
```

```python
mat_a.dot(mat_b)
```

```python
np.matmul(mat_a,mat_b)
```

```python
#eigen values, eigen vectors
lam,v = np.linalg.eig(mat_a)
```

```python
v[:,0] * lam[0] == mat_a @ v[:,0]
```
